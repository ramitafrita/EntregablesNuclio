INICIO DEL PROYECTO

```python
# Lectura de datos
import os
import pandas as pd
```

```python

import warnings
warnings.filterwarnings('ignore')
```

```python

df1 = pd.read_parquet('airbnb_madrid_conditions.parquet')
df2 = pd.read_parquet('airbnb_madrid_property.parquet')
df3 = pd.read_parquet('airbnb_madrid_location.parquet')
df4 = pd.read_parquet('airbnb_madrid_host.parquet')
df5 = pd.read_parquet('airbnb_madrid_reviews.parquet')

print(df1.shape, df2.shape, df3.shape, df4.shape, df5.shape)
```

```python
#unión de parquets:

for df in (df1, df2, df3, df4, df5):
    if '__index_level_0__' in df.columns:
        df.drop(columns='__index_level_0__', inplace=True)

df = df1.merge(
        right=df2,
        on='id'
    ).merge(
        right=df3,
        on='id'
    ).merge(
        right=df4,
        on='id'
    ).merge(
        right=df5,
        on='id'
    )

print('Filas y columnas del dataset:', df.shape)
df
```

```python
# Donde hay nulos? Miro si los datos están limpios para las hipotesis que me planteo.

df.isna().sum()

# En reviews hay nulos, lo tengo en cuenta por si necesito sacar información de esto para la hipotesis que haga.
```

```python
# Eliminar outliers: me quedo solo con precios menores al percentil 90 (saco el 10% que esté fuera)
df_filtrado = df[df['price'] < df['price'].quantile(0.90)]
```

HIPOTESIS 1 - ¿La ubicación influye en el precio?

```python
# Pienso que cuanto más cerca del centro (menor dist_km_sol), mayor el precio.

# Variables para identificar el dato:
# dist_km_sol: distancia a Puerta del Sol
# price: precio de la habitacion por noche

import plotly.express as px

fig = px.scatter(
    df_filtrado,
    x='dist_km_sol',
    y='price',
    opacity=0.4,
    title='Precio vs. Distancia al centro (Puerta del Sol)',
    labels={'dist_km_sol': 'Distancia al centro (km)', 'price': 'Precio por noche (€)'}
)

fig.update_layout(
    yaxis_title='Precio por noche (€)',
    xaxis_title='Distancia al centro (km)',
    template='plotly_white'
)

fig.show()
```

```python
# Correlacion:

df_filtrado[['dist_km_sol', 'price']].corr()

# Se observa una correlación negativa entre el precio y la distancia al centro que está en -0.2.
# Cuanto más cerca está un alojamiento de la Puerta del Sol, más alto tiende a ser el precio.
# Esto es coherente con la lógica del mercado: la ubicación céntrica aporta más valor.
```

HIPOTESIS 2 - ¿El tipo de alojamiento influye en el precio?

```python
# Creo que los alojamientos del tipo “entire home/apartment” son más caros que los “private room” o “shared room”.

# Variables clave que elijo:
# room_type: tipo de alojamiento
# price: precio por noche

# Hago un boxplot para comparar distribuciones entre categorías.

import plotly.express as px

fig = px.box(
    df_filtrado,
    x='room_type',
    y='price',
    points='all',  
    title='Precio según tipo de alojamiento',
    labels={'room_type': 'Tipo de alojamiento', 'price': 'Precio por noche (€)'}
)

fig.update_layout(
    yaxis_title='Precio por noche (€)',
    xaxis_title='Tipo de alojamiento',
    template='plotly_white'
)

fig.show()

df_filtrado.groupby('room_type')['price'].mean().sort_values(ascending=False)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtro de nuevo el df_filtrado por precio para tener una visión más clara, HAGO UN KDEPLOT.
df_kdeplot = df_filtrado[df_filtrado['price'] < 300]

plt.figure(figsize=(10,6))

# Apartamento completo
sns.kdeplot(
    data=df_kdeplot[df_kdeplot['room_type'] == 'entire_home_apt'], 
    x='price', 
    label='Apartamento entero',
    fill=True,
    alpha=0.5
)

# Habitación privada
sns.kdeplot(
    data=df_kdeplot[df_kdeplot['room_type'] == 'private_room'], 
    x='price', 
    label='Habitación privada',
    fill=True,
    alpha=0.5
)

# Habitación compartida
sns.kdeplot(
    data=df_kdeplot[df_kdeplot['room_type'] == 'shared_room'], 
    x='price', 
    label='Habitación compartida',
    fill=True,
    alpha=0.5
)

# Habitación de hotel
sns.kdeplot(
    data=df_kdeplot[df_kdeplot['room_type'] == 'hotel_room'], 
    x='price', 
    label='Habitación de hotel',
    fill=True,
    alpha=0.5
)

plt.title('Distribución de precios por tipo de alojamiento')
plt.xlabel('Precio por noche (€)')
plt.ylabel('Densidad de alojamientos')
plt.legend()
plt.grid(True)
plt.xlim(0, 300)
plt.show()
```

```python
# El tipo de alojamiento influye significativamente en el precio.
# Los alojamientos de tipo “Entire home/apartment” tienen, en promedio, precios más altos que las habitaciones privadas o compartidas.
# Hay muchas más densidad de habitaciones privadas o compartidas de precio economico que de apartamentos enteros o habitaciones de hotel, cosa que es lógica.
# Esto es esperable, ya que ofrecen más privacidad y espacio, y por tanto justifican un precio superior.
```

HIPOTESIS 3 - ¿Los alojamientos con más comodidades tienen precios más altos?

```python
# Pienso que cuantas más comodidades (Wifi, aire acondicionado, cocina, calefacción…), mayor es el precio del alojamiento.

# Variables de esta hipotesis: 
# amenities_wifi_internet
# amenities_kitchen
# amenities_air_conditioning
# amenities_heating
# amenities_tv

df_filtrado = df[df['price'] < df['price'].quantile(0.99)]

# Sumar cuántas comodidades tiene cada alojamiento (de las seleccionadas)
df_filtrado['total_comodidades'] = (
    df_filtrado[['amenities_wifi_internet',
                 'amenities_kitchen',
                 'amenities_air_conditioning',
                 'amenities_heating',
                 'amenities_tv']].sum(axis=1)
)
```

```python
# Relación de precio según la cantidad de comodidades que hay en el alojamiento:

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.stripplot(
    data=df_filtrado,
    x='total_comodidades',
    y='price',
    alpha=0.3      # Para que los puntos sean más transparentes
)
plt.title('Precio por número de comodidades')
plt.xlabel('Cantidad de comodidades')
plt.ylabel('Precio por noche (€)')
plt.grid(True)
plt.show()
```

```python
# Cada una de esas columnas representa la cantidad de alojamientos que tiene X comodidades (1,2,3,4 o 5)

# Se observa una tendencia creciente: a mayor número de comodidades, mayor es el precio del alojamiento.
# Esto indica que invertir en equipamiento y servicios puede justificar un aumento de tarifa.
# Especialmente destacable en alojamientos que ofrecen 4 o más comodidades.
```

```python
# Quiero añadir esta info a las hipotesis para saber si los alojamientos con aire acondicionado tienen un mayor precio

import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar precios menores al percentil 90 para el aire acondicionado y quitar outliers
df_filtrado_ac = df[df['price'] < df['price'].quantile(0.90)]

# Gráfico
plt.figure(figsize=(7,5))
sns.boxplot(
    data=df_filtrado_ac,
    x='amenities_air_conditioning',
    y='price',
    palette='coolwarm'
)
plt.title('¿Tener aire acondicionado impacta en el precio?')
plt.xlabel('¿Tiene aire acondicionado? (0=No, 1=Sí)')
plt.ylabel('Precio por noche (€)')
plt.grid(True)
plt.show()
```

HIPOTESIS 4 - ¿La capacidad del alojamiento influye en el precio?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creo franjas de capacidad en las habitaciones
def agrupar_capacidad(x):
    if x <= 2:
        return '1-2 personas'
    elif x <= 4:
        return '3-4 personas'
    elif x <= 6:
        return '5-6 personas'
    elif x <= 8:
        return '7-8 personas'
    else:
        return '9+ personas'

df['capacidad_franjas'] = df['accommodates'].apply(agrupar_capacidad)

# Filtro precios menores al percentil 90 para quitar algunos outliers de "df" y ver el gráfico mejor
filtro_precio = df['price'] < df['price'].quantile(0.90)
df_hipotesis4 = df[filtro_precio]

# Hago el boxplot con el DataFrame filtrado y las franjas ya creadas
plt.figure(figsize=(10,6))
sns.boxplot(
    data=df_hipotesis4,
    x='capacidad_franjas',
    y='price',
    palette='pastel',
    order=['1-2 personas', '3-4 personas', '5-6 personas', '7-8 personas', '9+ personas']
)
plt.title('Precio por noche según capacidad del alojamiento')
plt.xlabel('Capacidad del alojamiento')
plt.ylabel('Precio por noche (€)')
plt.grid(True)
plt.show()
```

HIPOTESIS 5 - ¿Puede el barrio influenciar significativamente en el precio?

```python
# Creo que el barrio de Madrid puede ser clave para identificar qué precios hay según la zona de la ciudad y comprobar si tiene relación con la distancia al centro.
```

```python
# Identifico los barrios del DataFrame

df_filtrado['neighbourhood'].unique()
```

```python
# Veo cuales son los barrios TOP haciendo sort values

# Agrupamos por barrio
df_barrio = (
    df_filtrado
    .groupby('neighbourhood')
    .agg(
        precio_medio=('price', 'mean'),
        cantidad=('id', 'count'),
        latitud=('latitude', 'mean'),
        longitud=('longitude', 'mean')
    )
    .reset_index()
)

# Nos quedamos con los 10 barrios más caros
df_top_barrios = df_barrio.sort_values(by='precio_medio', ascending=False)
df_top_barrios
```

```python
# Hago un mapa de Madrid para ser claro con los barrios más caros:

# Variables que necesitamos:
# Una columna con latitud (latitude)
# Una columna con longitud (longitude)
# Una columna con el barrio (neighbourhood)
# El precio por anuncio

# Con eso, agrupamos por barrio y calculamos:
# Precio medio
# Coordenadas promedio (para ubicar la “bola” del barrio)

import plotly.express as px

fig = px.scatter_mapbox(
    df_top_barrios,
    lat='latitud',
    lon='longitud',
    size='precio_medio',
    color='precio_medio',
    text='neighbourhood',
    hover_name='neighbourhood',
    hover_data={'precio_medio': ':.2f', 'cantidad': True},
    size_max=20,
    zoom=11,
    mapbox_style='carto-positron'
)

fig.update_layout(title='Barrios por precios de Madrid')
fig.show()
```

```python
# El análisis confirma que el barrio donde se ubica el alojamiento tiene un impacto directo en el precio medio por noche y que si está en el centro puede tener mayor precio pero no siempre es así.
# Hay barrios fuera del centro con un alto precio, quizás por ser zonas altas.
# A través del data frame y el mapa interactivo, se observa que zonas presentan precios medios notablemente más altos que otros barrios de Madrid.
# Esto indica que, más allá de la distancia al centro o de las características del alojamiento, la percepción y 
# demanda del barrio es un factor clave a la hora de fijar precios.
# Esta información permite al cliente ajustar las tarifas según el valor de mercado de cada zona, maximizando ingresos sin perder competitividad.
```

HIPOTESIS 6 - ¿Algunos barrios ofrecen mejor relación calidad/precio y pueden ser zonas de oportunidad para inversión?


```python
# ¿Cómo planteamos donde invertir?

# Quiero identificar barrios que tengan:
# Precio medio bajo o moderado
# Puntuación alta en valoración de ubicación (review_scores_location)
# Posible tendencia al alza (zonas en crecimiento, buena percepción sin ser caras)

# Agrupamos por barrio y calculamos

df_analisis_barrio = (
    df_filtrado
    .groupby('neighbourhood')
    .agg(
        precio_medio=('price', 'mean'),
        puntuacion_ubicacion=('review_scores_location', 'mean'),
        cantidad=('id', 'count')
    )
    .reset_index()
)

#Creamos un gráfico para ver calidad/precio (y/x)

import plotly.express as px

fig = px.scatter(
    df_analisis_barrio,
    x='precio_medio',
    y='puntuacion_ubicacion',
    size='cantidad',
    text='neighbourhood',  # Nombres de barrios en el gráfico
    hover_name='neighbourhood',
    hover_data={'precio_medio': ':.2f', 'puntuacion_ubicacion': ':.2f', 'cantidad': True}, #Ver datos en los globos del gráfico interactivo
    title='Relación calidad/precio por barrio',
    size_max=40
)

# Mejoras visuales del gráfico
fig.update_traces(marker=dict(opacity=0.6, line=dict(width=1, color='DarkSlateGrey')))
fig.update_layout(
    xaxis_title='Precio medio (€)',
    yaxis_title='Puntuación media de ubicación',
    template='plotly_white'
)

fig.show()
```

```python
#El análisis de la relación entre el precio medio y la puntuación media de ubicación por 
#barrio revela diferencias significativas que pueden guiar decisiones de inversión estratégicas.

#En el gráfico interactivo, cada burbuja representa un barrio de Madrid:
#El eje X indica el precio medio por noche.
#El eje Y indica la valoración media de la ubicación que hacen los huéspedes.
#El nombre del barrio aparece como etiqueta y también al pasar el ratón.

#Los barrios que aparecen en la zona superior izquierda del gráfico, es decir, alta puntuación en ubicación y 
#precio medio bajo o moderado, representan zonas de oportunidad de inversión.

#Estos barrios combinan una buena percepción en valoraciones con un valor de mercado aún accesible en comparación a otros barrios,
#lo que sugiere que podrían experimentar una revalorización futura o probablemente permitan obtener más rendimiento por euro invertido.
```

```python
# Filtro de barrios con buena puntuación (mayor a 9) y precio medio moderado
oportunidades_inversion = df_analisis_barrio[
    (df_analisis_barrio['puntuacion_ubicacion'] > 9) &
    (df_analisis_barrio['precio_medio'] < 100)
].sort_values(by=['puntuacion_ubicacion', 'precio_medio'], ascending=[False, True])

# Mostramos los 5 principales
oportunidades_inversion.head(5)
```
